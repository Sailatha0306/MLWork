```python
l = [('cat','good'),('dog','bad'),('goat','good')]
```

```python
cfd = ConditionalFreqDist(l)
```

```python
cfd
```

```python
cfd = ConditionalFreqDist((genre,word)
                          for genre in brown.categories()
                          for word in brown.words(categories = genre))
```

```python
genre = ['news','religion','hobbies','romance','humor']
modals = ['can','could','may','might','must']
cfd.tabulate(conditions = genre,samples = modals)
```
